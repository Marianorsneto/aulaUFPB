```python
#resolução do desafio 2.3
#dividindo os valores pela metragem
dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
```

```python
#criar nova coluna no dateframe
dados_vendas["Valor_m2"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
dados_vendas.head()
```

```python
#agrupando pelos bairros chamando uma função
dados_vendas.groupby("Bairro").mean()
```

```python
#soma da metragem e valor anuncio para os bairros
dados_bairro = dados_vendas.groupby("Bairro").sum()
dados_bairro
```

```python
#criando mais uma coluna no dataframe
dados_bairro["Valor_m2_bairro"] = dados_bairro["Valor_anuncio_float"]/dados_bairro["Metragem"]
dados_bairro
```

```python
#comparativo dos dois metodos de análise
dados_vendas.groupby("Bairro").mean().loc["Vila Mariana"]
```

```python
dados_bairro.loc["Vila Mariana"]
```

```python
dados_bairro["Valor_m2_bairro"]["Vila Mariana"]
```

```python
#descobrindo os bairros com mais anuncios
dados_vendas["Bairro"].value_counts()
```

```python
dados_vendas["Bairro"].value_counts().index
```

```python
dados_vendas["Bairro"].value_counts()[:10].index
```

```python
top_bairros = dados_vendas["Bairro"].value_counts()[:10].index
```

```python
#alterando o bairro de index para coluna, associar ao dataset , utilizando o parametro inplace
dados_bairro.reset_index(inplace = True)
dados_bairro
```

```python
#plotando os 10 top bairros, utiliza-se o @ para lincar com a variavel top bairros
dados_bairro.query("Bairro in @top_bairros")
```

```python
#buscar no google seaborn bar plot para verificar exemplos possiveis
plt.figure(figsize=(10, 8))
ax = sns.barplot(x="Bairro", y="Valor_m2_bairro", data=dados_bairro.query("Bairro in @top_bairros"))
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros"), x="Bairro", y="Valor_anuncio_float")
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
#analisando as metragens menores de 30000, incluindo uma condição através do &
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros & Metragem < 30000"), x="Bairro", y="Metragem")
ax.tick_params(axis='x', rotation=45)
plt.show()
```
